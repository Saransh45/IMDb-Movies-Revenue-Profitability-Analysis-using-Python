### Profit Analysis
Profit was calculated as Revenue minus Budget for each movie. This metric helps
identify which movies were commercially successful, beyond just high revenue.

```python
def calculate_profit(row):
  return row['Revenue'] - row['budget']

data.loc[:,'profit'] = data.apply(calculate_profit,axis=1)
```

```python
def revised_profit(revenue,budget):
  return revenue-budget

data['new_profit'] = data.apply(lambda x: revised_profit(x['Revenue'], x['budget']), axis=1)
```

```python
data.loc[:,'profit_value'] = data.loc[:,'profit'].apply(lambda x: 'positive' if x>0 else 'negative')
```

```python
data.loc[:,'original_language'] = data.loc[:,'original_language'].apply(lambda x: x.upper())
```

```python
data.loc[:,'popularity_new'] = data.loc[:,'popularity'].apply(lambda x: round(x,2))
```

```python
import ast

def func(value, key):
    try:
        j = ast.literal_eval(value)
        return ",".join([i[key] for i in j])
    except:
        return np.nan
```

Genres are stored as JSON-like strings and were parsed to extract genre names.

```python
data.loc[:,'genre_names'] = data.loc[:,'genres'].apply(func, key='name')
```

```python
data.loc[:,'release_date'] = pd.to_datetime(data.loc[:,'release_date'])
```

```python
data.loc[:,'release_year'] = pd.to_datetime(data.loc[:,'release_date']).dt.year
```

#GroupBy Insights

```python
data = data[data['release_year']<=2017].copy()
data.groupby('release_year')['Revenue'].sum().sort_index(ascending = False)
```

### Revenue Trend by Release Year

Total movie revenue shows a clear growth pattern from the late 1990s,
with a strong surge between 2009 and 2015. The highest total revenue
was observed in 2015, followed by consistently high revenues until 2016.
A noticeable decline is seen in 2017, which may be influenced by fewer
movies or incomplete data for that year.

```python
data.groupby('original_language')['Revenue'].sum().sort_values(ascending=False)
```

### Revenue by Original Language

**English**-language movies overwhelmingly dominate total revenue,
contributing the vast majority of box-office earnings. Other languages
such as Japanese (JA), Chinese (ZH), and Swedish (SV) generate
significantly lower revenue in comparison. This highlights the strong
global reach and commercial dominance of English-language films.

```python
data.groupby('release_year')[['Revenue','budget']].sum().sort_index(ascending = False)
```

### Revenue vs Budget Trend by Release Year

Both movie budgets and revenues show a steady increase over time,
indicating higher investment and scale in the film industry. From the
early 2000s onward, revenue growth generally outpaces budget growth,
suggesting improved profitability during peak years such as 2009, 2012,
and 2015. Although budgets continue to rise, revenue fluctuations in
later years indicate that higher investment does not always guarantee
proportionally higher returns.

```python
data.groupby('release_year')['runtime'].max().reset_index()
```

### Maximum Movie Runtime by Release Year

The maximum runtime of movies shows a gradual increase over time,
indicating a trend toward longer feature films in later years. Earlier
years (1970s–1980s) generally had maximum runtimes around 120–130 minutes,
whereas from the 2000s onward, several years feature movies exceeding
150 minutes. This suggests growing audience acceptance of longer films,
often associated with epic storytelling, franchises, or high-budget
productions.
